记录了 7个关节的角度信息，获取两个循环里的所有信息，循环区间读取待优化

```python
import numpy as np
import matplotlib.pyplot as plt
import os
# 加载 .npy 文件
file_path = '/home/shifei/code/isaacgyms/arm_dof_pos_cumulative.npy'  # 替换为实际的 .npy 文件路径
data = np.load(file_path)

# 查看数据
data
joint_data = np.load(file_path)

# 获取关节数量
num_dofs = joint_data.shape[1]

# 创建保存图像的文件夹
output_folder = 'joint_images'
os.makedirs(output_folder, exist_ok=True)

# 遍历每个关节
for dof_idx in range(num_dofs):
    # 提取当前关节的数据
    joint_angle = joint_data[:,dof_idx]  # 假设只取第一个环境的数据

    # 创建一个新的图形
    plt.figure()

    # 绘制关节角度随步数的变化曲线
    plt.plot(joint_angle)
    plt.title(f'Joint {dof_idx} Angle')
    plt.xlabel('Step')
    plt.ylabel('Angle')
    # plt.show()
    # 保存图像
    image_path = os.path.join(output_folder, f'joint_{dof_idx}_angle.png')
    plt.savefig(image_path)

    # 关闭图形
    # plt.close()

print(f'Images saved to {output_folder}')
```

夹爪的两个关节控制思路：这两幅图像指的是他们距离中心点的距离，0.05就是夹爪完全张开，这两幅图像下发的是状态信息，也就是在这中间要求夹爪能够完全夹住车标，那么就可以通过认为写死,暂时不采用写死的方法，而是采用直接进行透传模式的土方法来控制
